```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import warnings
from tqdm import tqdm
pd.set_option('display.max_columns', None)
warnings.filterwarnings('ignore')
```

```python
ratings = np.load('./../Content Based/Dataset-Numpy/ratings.npy')
print('Row : Users\nColumn : Movies')
ratings.shape
```

```python
def pearson(tmp_a, tmp_b):
    a = tmp_a
    b = tmp_b
    mean_a = 0
    mean_b = 0
    count_a = 0
    count_b = 0
    norm_a = 0
    norm_b = 0
    
    for i in range(len(a)):
        if not np.isnan(a[i]):
            count_a += 1
            mean_a += a[i]
        if not np.isnan(b[i]):
            count_b += 1
            mean_b += b[i]
    mean_a = mean_a / count_a
    mean_b = mean_b / count_b
    
    for i in range(len(a)):
        if not np.isnan(a[i]):
            a[i] = a[i] - mean_a
        if not np.isnan(b[i]):
            b[i] = b[i] - mean_b
    for i in range(len(a)):
        
        if np.isnan(a[i]):
            a[i] = 0
        if np.isnan(b[i]):
            b[i] = 0
    dot = np.dot(a, b)
    for i in range(len(a)):
        norm_a = norm_a + np.square(a[i])
        norm_b = norm_b + np.square(b[i])
        
    norm_a = np.sqrt(norm_a)
    norm_b = np.sqrt(norm_b)
    
    return dot / (norm_a * norm_b)
```

### Finding Most Similar User to User 1

```python
numUsers = ratings.shape[0]
user = 0
```

```python
similarities = []
for i in tqdm(range(numUsers)):
    
    if i == user:
        continue
    
    sim = pearson(ratings[i], ratings[user])
    similarities.append(sim)
most_similar = np.argmax(np.asarray(similarities))
```

```python
print('Most Similar User to {} is User {}'.format(user+1, most_similar))
print('Similarity : {}'.format(similarities[most_similar]))
```
